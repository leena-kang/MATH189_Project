## Test of Homogeneity

### Mother's and Father's qualification

> Is there a difference in the distribution of the each students' parent's qualification between those who dropped out and those who graduated/are enrolled?

#### Mother's Qualification

```python
pivot_table = pd.crosstab(df1['Target'], df1['Mother\'s qualification'])
pivot_table
```

```python
sqtab = sm.stats.SquareTable(pivot_table)
print(sqtab.homogeneity())
```
